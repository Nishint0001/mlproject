```python
#define numerical & categorical columns
numeric_features=[ng for ng in df.columns if df[ng].dtype!='O']
categorical_features=[ng for ng in df.columns if df[ng].dtype=='O']


#print values

print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('We have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
#MAKE 2 new columns total score and avg

df["total score"]=df["math_score"]+df["reading_score"]+df["writing_score"]
df['average']=df["total score"]/3
df.head(3)
```

```python
r_full=df[df["reading_score"]==100]['average'].count()
print("Number of students with full marks in Reading: ",r_full)


w_full=df[df["writing_score"]==100]['average'].count()
print("Number of students with full marks in writing: ",w_full)


m_full=df[df["math_score"]==100]['average'].count()
print("Number of students with full marks in maths: ",m_full)
```
